```python
#Seeing the first plot of the target variable, it definetely needs to be scaled before being used on our ML model. Since it's very skewed, we decided to log-scale it
target = df_measurements.utime
to_log = lambda x:np.log(x+1) #If we ever get a 0, so that we get +1 
target = to_log(target)
```

```python
g = sn.displot(target)
g.fig.set_figwidth(16)
g.fig.set_figheight(10)
```

```python
df_measurements.utime = target
```

```python
X = df_measurements[df_measurements.columns[1:-1]]
Y = df_measurements[df_measurements.columns[-1]]
```

```python
X.select_dtypes(exclude = "number")
```

```python
X["codec"] = X["codec"].astype('category')
X["o_codec"] = X["o_codec"].astype('category')
X["codec_cat"] = X["codec"].cat.codes
X["ocodec_cat"] = X["o_codec"].cat.codes
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X.select_dtypes("number"),Y, test_size = 0.3, random_state = 123)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)          # Il ne faut fiter que sur les data d'entrainement
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)  # apply same transformation to test data
```

```python
import xgboost as xgb
algo = xgb.XGBRegressor(colsample_bytree=  0.5,learning_rate = 0.1, max_depth = 5, alpha = 100, n_estimators = 100)
```

```python
modele = algo.fit(X_train, Y_train)
modele.score(X_test, Y_test)
```

```python
modele.get_params()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint
params = {  
    "n_estimators": randint(10, 150),
    "max_depth": randint(3, 10),
    "learning_rate": uniform(0.05, 0.3),
    "colsample_bytree": uniform(0.7,0.3),
    "subsample": uniform(0.6, 0.3),
    "gamma": uniform(0, 10)
}
search = RandomizedSearchCV(modele, param_distributions=params, random_state=42, n_iter=200, cv=5, verbose=1, n_jobs=1, return_train_score=True)
```

```python
search.fit(X_train,Y_train)
```

```python
xgb.plot_importance(modele)
```

```python
preds = algo.predict(X_test)
pred_train = algo.predict(X_train)
rmse = np.sqrt(mean_squared_error(Y_test, preds))
rmse_train = np.sqrt(mean_squared_error(Y_train, pred_train))
r2 = r2_score(Y_test,preds)
r2_train = r2_score(Y_train,pred_train)
print("RMSE :",rmse_train,"\n R2 :",r2_train)
print("RMSE :",rmse,"\n R2 :",r2)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import learning_curve
cv = ShuffleSplit(n_splits=100, test_size=0.2, random_state=0)
```

```python
train_sizes, train_scores, test_scores, fit_times, _ = learning_curve(modele, np.concatenate([X_train,X_test],axis =0),np.concatenate([Y_train,Y_test],axis = 0), 
                                                                      cv=cv, n_jobs=4,train_sizes=np.linspace(.1, 1.0, 5),
                                                                      return_times=True)
```

```python
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)
fit_times_mean = np.mean(fit_times, axis=1)
fit_times_std = np.std(fit_times, axis=1)
```

```python
# Plot fit_time vs score
axes = plt.plot(fit_times_mean, test_scores_mean, 'o-')
plt.fill_between(fit_times_mean, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1)
plt.xlabel("fit_times")
plt.ylabel("Score")
plt.title("Performance of the model")
```

### Comparison with a Linear Regression model 

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression(normalize = True).fit(X_train, Y_train)
reg.score(X_train, Y_train)
```

```python
pred = reg.predict(X_test)
pred_train = reg.predict(X_train)
print("RMSE for Train Test: ",np.sqrt(mean_squared_error(Y_train,pred_train)))
print("RMSE for Test Test: ",np.sqrt(mean_squared_error(Y_test,pred)))
print("RÂ² train :", r2_score(Y_train,pred_train))
print("RÂ² test :",r2_score(Y_test,pred))
```
